```python
# importing modules
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# setting the base url & user agent
baseurl = "https://www.kilimall.co.ke/"
headers = {'User-Agent' : 'https://explore.whatismybrowser.com/useragents/parse/4396-aol-browser-windows-webkit'} 
```

```python
# getting data with product link
tv_links = []

for x in range(1,119):
    r = requests.get(f"https://www.kilimall.co.ke/category/televisions?backCid=5406&form=category&source=category|allCategory|televisions&page={x}")
    soup = BeautifulSoup(r.content, "html.parser")
    television_list = soup.find_all("div", class_ = "product-item")

    for television in television_list:
        for link in television.find_all("a", href = True):
            tv_links.append(baseurl + link["href"])

print(len(tv_links)) # there are 4240 product links from the television 
```

```python
# write the list to a txt file
with open ("tv_links.txt", "w") as file:
    for item in tv_links:
        file.write(f"{item}\n")
```

```python
# getting data from the link
#tv_link_txt = r"tv_links.txt"

#image_link = "https://www.kilimall.co.ke/listing/2517665-vitron-32-inch-frameless-smart-tv-hd-netflix-tv-htc3200s-youtube-television-dvbt2-android-11-ac-energy-saving-television-1g8g?skuId=18272411"

product_list_tv = []

for link in tv_links:
    print("saving:" + str(tv_links.index(link)))
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, "html.parser")

    product_id = soup.find("div", class_ = "last-b-name name").text.strip()
    name = soup.find("div", class_ = "product-title").text.strip()
    current_price = soup.find("span", class_ = "sale-price").text.strip()
    previous_price = soup.find("span", class_ = "del-price").text.strip()
    reviews = soup.find("span", class_ = "reviews").text.strip()
    try:
        rating = soup.find("span", class_ = "rate").text.strip()
    except:
        rating = "no rating"
    
    # define a dictionary
    tvs_dict = {
        'product_id': product_id,
        'name': name,
        "rating": rating,
        "reviews": reviews,
        "current_price": current_price,
        "previous_price": previous_price
    }

    product_list_tv.append(tvs_dict)
    
    # Wait for 5 seconds before the next request
    time.sleep(5)
```
