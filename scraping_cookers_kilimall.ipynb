## Scraping Gas Cookers Data from Kilimall

### 1. Importing Libraries

```python

import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### 2. Describing the base url and user agent variables

```python
baseurl = "https://www.kilimall.co.ke/"
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.6312.123 Safari/537.36 AOLShield/123.0.6312.3'} 
```

### 3. A loop that iterates through the website, scrapes data

#### (a) Cooktops data

```python
cooktops_list_no_rating = [] # initializing an empty list that will store the results of the for loop

for x in range(1,137): # range of website pages that has cooktops product displayed on them
    print("saving:" + str(x))
    r_cooktops = requests.get(f"https://www.kilimall.co.ke/category/cooktops?backCid=5461&form=category&source=search|enterSearch|Gas+Cookers&page={x}")
    soup_cooktops = BeautifulSoup(r_cooktops.content, "html.parser")
    cooktops_list = soup_cooktops.find_all("div", class_ = "info-box")
    #print(cooktops_list)
    for item in cooktops_list:
        cooktop_name = item.find("p", class_ = "product-title").text.strip()
        cooktop_price = item.find("div", class_ = "product-price").text.strip()
        cooktop_reviews = item.find("span", class_ = "reviews").text.strip()
        # define a dictionary
        cooktops_dict = {
            "cooktop_name": cooktop_name,
            "cooktop_reviews": cooktop_reviews,
            "cooktop_price": cooktop_price,
            }
        cooktops_list_no_rating.append(cooktops_dict)

print(len(cooktops_list_no_rating)) # checks the number of cooktops saved in this list
```

```python
# converting the list to a dataframe
cooktops_df = pd.DataFrame(cooktops_list_no_rating)
cooktops_df
```

```python
# save as csv
cooktops_df.to_csv("kilimall_cooktops_no_rating.csv", index = False, encoding = "utf-8")
```

#### (b) Standing Gas Cookers data

```python
standing_cookers = [] # initializing an empty list that will store the results of the for loop

for x in range(1, 33): # range of website pages that has cooktops product displayed on them
    print("saving:" + str(x))
    r_standing_cookers = requests.get(f"https://www.kilimall.co.ke/category/standing-gas-cookers?id=2207&form=category&page={x}", headers=headers)
    soup__standing_cookers = BeautifulSoup(r_standing_cookers.content, "html.parser")
    standing_cookers_list = soup__standing_cookers.find_all("div", class_ = "info-box")
    #print(len(standing_cookers_list))
    
    for item in standing_cookers_list:
        standing_cooker_name = item.find("p", class_ = "product-title").text.strip()
        standing_cooker_price = item.find("div", class_ = "product-price").text.strip()
        standing_cooker_reviews = item.find("span", class_ = "reviews").text.strip()
        # define a dictionary
        standing_cooker_dict = {
            "standing_cooker_name": standing_cooker_name,
            "standing_cooker_reviews": standing_cooker_reviews,
            "standing_cooker_price": standing_cooker_price,
            }
        standing_cookers.append(standing_cooker_dict)

print(len(standing_cookers)) # checks the number of cooktops saved in this list
```

```python
# converting the list to a dataframe
standing_cooker_df = pd.DataFrame(standing_cookers)
standing_cooker_df
```

```python
# save as csv
standing_cooker_df.to_csv("kilimall_standing_cooker_no_rating.csv", index = False, encoding = "utf-8")
```
