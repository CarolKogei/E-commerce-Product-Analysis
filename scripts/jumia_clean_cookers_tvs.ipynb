```python
# import necessary modules
import numpy as np
import pandas as pd
import datetime
```

```python
cookers = pd.read_csv('/workspaces/e-commerce-product-analysis/misc/data/scrapped/jumia_cookers.csv', index_col=0)
cookers.reset_index(inplace=True)
cookers.index = cookers.index + 1
cookers.rename(columns={'index': 'id'}, inplace=True)
```

```python
cookers.info()
```

```python
cookers = cookers.drop(cookers.columns[[14, 15]], axis=1)
cookers.head(3)
#cookers.info()
```

```python
cookers = cookers.astype({
    'name': 'string',
    'item_id': 'string',
    'category': 'string',
    'subcategory': 'string',
    'subcategory2': 'string',
    'subcategory3': 'string',
    'brand': 'string',
    'source': 'string',
    'url': 'string'
 
})
```

```python
cookers['id'] = range(1, len(cookers) + 1)
```

```python
cookers.head()
```

```python
cookers.info()
```

```python
cookers['id'] = cookers.iloc[:, 0]
cookers['id'] = cookers['id'] + 1
cookers.head(5)
```

```python
cookers.describe()

duplicates = cookers[cookers.duplicated()]
duplicates.count()
#len(cookers)
# cookers.isna().sum()
```

```python
# Drop duplicates based on the 'id' column, keeping the first occurrence
cookers = cookers.drop_duplicates(subset='id', keep='last')

# Save the cleaned DataFrame to a new CSV file
cookers.to_csv('cleaned_jumia_scraped_televisions.csv', index=False)

cookers.head(100)
cookers.isna().sum()
```

```python
# Replace NaN values in 'previous_price' with values from 'discounted_price'
cookers['previous_price'] = cookers['previous_price'].fillna(cookers['discounted_price'])

# Replace NaN values in 'discount_%' and 'reviews_count' with 0
cookers[['discount_%', 'reviews_count']] = cookers[['discount_%', 'reviews_count']].fillna(0)
```

For all previous extracted as null,   `previous_price == discounted_price equal`  For all products with 0 reviews assign None rating, all null review counts

```python
# Replace NaN values in 'reviews_count and ratings' with 0
cookers[['reviews_count', 'rating']] = cookers[['reviews_count', 'rating']].fillna(0)
cookers['rating'] = cookers['rating'].fillna("None")

cookers.head(100)
#cookers.isna().sum()
```
